```python
# Analisandop a coluna Elenco
# A coluna 'elenco' contém strings de atores separados por vírgulas.
# Primeiro, vamos preencher os valores 'Não informado' com uma string vazia para facilitar o split.
df['elenco'] = df['elenco'].replace('Não informado', '')

# Dividir as strings de 'elenco' em listas de atores e expandir para linhas separadas.
# Em seguida, remover espaços em branco de cada nome e filtrar nomes vazios.

elenco_separado = df['elenco'].str.split(', ').explode()
elenco_limpo = elenco_separado[elenco_separado != ''].str.strip()

# Contar a frequência de cada ator
atore_counts = elenco_limpo.value_counts()

print("Top 5 Atores/Atrizes com Mais Títulos:")
display(atore_counts.head(5))
```

```python
#Top 5 atores

# Obter os top 5 atores/atrizes para visualização
top_5_atores = atore_counts.head(5)

plt.figure(figsize=(10, 6))
sns.barplot(x=top_5_atores.values,
            y=top_5_atores.index,
            legend=False)
plt.title('Top 5 Atores/Atrizes com Mais Títulos')
plt.xlabel('Número de Títulos')
plt.ylabel('Ator/Atriz')
plt.show()
```

```python
import numpy as np

# Substituir 'Não informado' por NaN na coluna 'ano_adicao' para permitir conversão para int
df['ano_adicao'] = df['ano_adicao'].replace('Não informado', np.nan)

# Converter 'ano_adicao' para o tipo inteiro (Int64 para suportar valores nulos)
df['ano_adicao'] = df['ano_adicao'].astype('Int64')

print("Verificando o tipo de dado da coluna 'ano_adicao' após a conversão:")
display(df['ano_adicao'].head())
display(df['ano_adicao'].dtype)
```

### Visualização dos Top 10 Anos com Mais Títulos Adicionados

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Obter os 10 anos com mais adições
top_10_adicoes = df['ano_adicao'].value_counts().head(10)

plt.figure(figsize=(12, 6))
sns.countplot(data=df,
              x='ano_adicao',
              order=top_10_adicoes.index.astype(str),
              palette='crest',
              hue='ano_adicao',
              legend=False)
plt.title('Top 10 Anos com Mais Títulos Adicionados')
plt.xlabel('Ano de Adição')
plt.ylabel('Número de Títulos')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
# 1.lista com os NOMES dos top 5 países
lista_top_5_paises = top_10.index.tolist()

# 2. Dataframe filtradopara pegar SÓ filmes desses 5 países
df_filtrado = df[df['pais'].isin(lista_top_5_paises)].copy()

# 3. Limpeza e Explode (Necessário para separar os atores mantendo o país)
df_filtrado['elenco'] = df_filtrado['elenco'].replace('Não informado', '')

# Criação de uma linha para cada ator, mas mantendo a coluna 'pais' ao lado!
df_atores_pais = df_filtrado.assign(ator=df_filtrado['elenco'].str.split(', ')).explode('ator')

# Limpeza de espaços vazios e espaços em branco extras
df_atores_pais = df_atores_pais[df_atores_pais['ator'] != '']
df_atores_pais['ator'] = df_atores_pais['ator'].str.strip()

# 4. Agrupar por País E Ator
# Conta quantas vezes cada ator aparece em cada país
contagem = df_atores_pais.groupby(['pais', 'ator']).size().reset_index(name='quantidade_filmes')

# 5. Filtrar o top 1 de cada país
# O idxmax() descobre o índice da linha que tem o maior valor para cada país
indices_campeoes = contagem.groupby('pais')['quantidade_filmes'].idxmax()

# Seleção das linhas vencedoras
tabela_atores_top5 = contagem.loc[indices_campeoes].sort_values('quantidade_filmes', ascending=False)

# 6. Exibindo a Tabela Final
print("Ator/Atriz mais frequente em cada um dos Top 5 Países:")
display(tabela_atores_top5)
```

```python
#Top 5 atores de cada país

plt.figure(figsize=(10, 6))

sns.barplot(x=tabela_atores_top5['quantidade_filmes'],
            y=tabela_atores_top5['ator'],
            hue=tabela_atores_top5['pais'],
            dodge=False)
plt.title('Ator/Atriz Principal Top 5 Países')
plt.xlabel('Número de Títulos')
plt.ylabel('Ator Principal')
plt.legend(title='País')
plt.show()
```
